```python
# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of 
# Matplotlib.
# Answer :-
# Matplotlib is a popular data visualization library in Python. It provides a wide range of tools for creating various types of plots and graphs. Matplotlib is widely used in the scientific and data analysis communities for its flexibility and ease of use.

# Here are five plots that can be created using the Pyplot module of Matplotlib:

# Line Plot: It represents data as points connected by straight lines. It is commonly used to visualize trends or relationships between two variables.

# Bar Plot: It uses rectangular bars to compare categories or groups of data. It is useful for displaying and comparing discrete or categorical data.

# Histogram: It displays the distribution of a continuous variable. It divides the data into bins and shows the frequency or count of values falling within each bin.

# Scatter Plot: It shows the relationship between two variables by displaying points on a two-dimensional coordinate system. It is useful for visualizing correlations or clusters in data.

# Pie Chart: It represents data as sectors of a circle, where each sector represents a different category or proportion of a whole. It is commonly used to show the composition or distribution of a dataset.

# These are just a few examples of the plots that can be created using the Pyplot module of Matplotlib. Matplotlib offers many other types of plots and customization options to meet different visualization needs.
```

```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
# plot a scatter plot.
#  import numpy as np
#  np.random.seed(3)
#  x = 3 + np.random.normal(0, 2, 50)
#  y = 3 + np.random.normal(0, 2, len(x))
#  Note: Also add title, xlabel, and ylabel to the plot.
# Answer :-
# A scatter plot is a type of plot that displays the relationship between two variables as points on a two-dimensional coordinate system. Each point represents an observation or data point, with the x-coordinate corresponding to one variable and the y-coordinate corresponding to another variable. Scatter plots are useful for visualizing correlations, clusters, or patterns in data.
# Here's the code to generate data for x and y and create a scatter plot using Matplotlib:
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Explanation of the code:

# import numpy as np: This imports the NumPy library, which provides functions for generating random numbers.
# import matplotlib.pyplot as plt: This imports the Pyplot module from the Matplotlib library, which is used for creating plots.
# np.random.seed(3): This sets a seed for the random number generator to ensure reproducibility.
# x = 3 + np.random.normal(0, 2, 50): This generates an array of 50 random numbers with a mean of 3 and a standard deviation of 2.
# y = 3 + np.random.normal(0, 2, len(x)): This generates an array of random numbers with the same length as x, following the same mean and standard deviation.
# plt.scatter(x, y): This creates a scatter plot with x as the x-coordinates and y as the y-coordinates.
# plt.title("Scatter Plot"): This sets the title of the plot as "Scatter Plot".
# plt.xlabel("X"): This sets the label for the x-axis as "X".
# plt.ylabel("Y"): This sets the label for the y-axis as "Y".
# plt.show(): This displays the scatter plot.
# Running this code will generate a scatter plot with the given data, along with a title, x-axis label, and y-axis label.
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
#  Use the following data:
#  import numpy as np
#  For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])
#  For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
#  For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
#  import numpy as np
#  For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
# Answer :-
import numpy as np
import matplotlib.pyplot as plt

# Data for Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line 4")

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
#  import numpy as np
# company = np.array(["Apple", "Microsoft", "Google", "AMD"])
#  profit = np.array([3000, 8000, 1000, 10000])
# Answer :-
# A bar plot, also known as a bar chart or bar graph, is a type of plot that uses rectangular bars to represent data. Each bar corresponds to a category or group, and the height or length of the bar represents the value or frequency of that category. Bar plots are commonly used to compare and display categorical data.

# Bar plots are used for several purposes:

# Comparison: Bar plots allow easy comparison of values across different categories or groups. They provide a visual representation of the magnitude or frequency of the data points.
# Categorical Data: Bar plots are effective for visualizing categorical data and showing the distribution or composition of each category.
# Trend Analysis: Bar plots can be used to analyze trends over time or different conditions. By plotting data for different time periods or conditions, it becomes easy to observe patterns or changes.
# Now let's create a bar plot and a horizontal bar plot using the given data:
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(8, 5))  # Adjust the figure size (optional)
plt.bar(company, profit)
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 5))  # Adjust the figure size (optional)
plt.barh(company, profit)
plt.title("Company Profits")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
# Explanation of the code:

# import numpy as np: This imports the NumPy library.

# import matplotlib.pyplot as plt: This imports the Pyplot module from the Matplotlib library.

# Two NumPy arrays, company and profit, are created to represent the company names and their corresponding profits.

# Bar Plot:

# plt.bar(company, profit): This creates a vertical bar plot using the company array as the x-values and the profit array as the heights of the bars.
# plt.title("Company Profits"): This sets the title of the bar plot as "Company Profits".
# plt.xlabel("Company"): This sets the label for the x-axis as "Company".
# plt.ylabel("Profit"): This sets the label for the y-axis as "Profit".
# plt.show(): This displays the bar plot.
# Horizontal Bar Plot:

# plt.barh(company, profit): This creates a horizontal bar plot using the company array as the y-values and the profit array as the lengths of the bars.
# plt.title("Company Profits"): This sets the title of the horizontal bar plot as "Company Profits".
# plt.xlabel("Profit"): This sets the label for the x-axis as "Profit".
# plt.ylabel("Company"): This sets the label for the y-axis as "Company".
# plt.show(): This displays the horizontal bar plot.
# Running this code will generate a bar plot and a horizontal bar plot using the given data, with appropriate labels and titles. The bar plot will display the profits for each company as vertical bars, and the horizontal bar plot will show the same information with horizontal bars.
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
#  box1 = np.random.normal(100, 10, 200)
#  box2 = np.random.normal(90, 20, 200)
# Answer :-
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Boxes")
plt.ylabel("Values")
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()

```
