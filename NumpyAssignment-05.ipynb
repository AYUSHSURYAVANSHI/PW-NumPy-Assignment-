```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
# Answer :-
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare your data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Set up the plot
p = figure(title='My Bokeh Plot', x_axis_label='X', y_axis_label='Y')

# Add glyphs to the plot
p.circle(x, y, size=10, color='blue', alpha=0.5)  # Add scatter plot with circles
p.line(x, y, line_width=2, color='red')  # Add line plot

# Display the plot
output_notebook()
show(p)

```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
# Answer :-
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Legend

# Prepare your data
x = [1, 2, 3, 4, 5]
y1 = [6, 7, 2, 4, 5]
y2 = [3, 2, 5, 6, 1]

# Set up the plot
p = figure(title='My Bokeh Plot', x_axis_label='X', y_axis_label='Y')

# Add glyphs to the plot
p.circle(x, y1, size=10, color='blue', alpha=0.5, legend_label='Data 1')
p.line(x, y2, line_width=2, color='red', legend_label='Data 2')

# Customize plot appearance
p.title.text_font_size = '18pt'  # Set title font size
p.xaxis.axis_label_text_font_style = 'italic'  # Set x-axis label font style
p.yaxis.axis_label_text_color = 'green'  # Set y-axis label text color

# Customize axis tick labels
p.xaxis.major_label_text_color = 'orange'  # Set x-axis tick label text color
p.yaxis.major_label_text_font_size = '12pt'  # Set y-axis tick label font size

# Customize legend
p.legend.location = 'top_left'  # Set legend position
p.legend.label_text_font_style = 'bold'  # Set legend label font style

# Display the plot
output_notebook()
show(p)
```
